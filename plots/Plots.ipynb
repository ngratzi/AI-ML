# Load Data Set

```python
filename = '../../data/celebrity-heights.csv'
df = pd.read_csv(filename)
```

# Histograms

## Matplotlib

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

```python
plt.hist(df['inches'], color='skyblue')
plt.xlabel('Height in Inches')
plt.ylabel('Frequency')
plt.title('Matplotlib Histogram')
# plt.savefig('matplotlib_histogram.png')
```

## Seaborn

https://seaborn.pydata.org/generated/seaborn.histplot.html

```python
# plt.figure(figsize=(15,8))
sns.histplot(df['inches'], color='purple')
plt.title('Seaborn Histogram')
plt.xlabel('Height in Inches')
plt.ylabel('Frequency')
# plt.savefig('seaborn_histogram.png')
```

## Pandas

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.hist.html

```python
df['inches'].plot(kind='hist', color='orange')
plt.title('Pandas Histogram')
plt.xlabel('Height in Inches')
plt.ylabel('Frequency')
# plt.savefig('pandas_histogram.png')
```

## Plotly

```python
fig = px.histogram(df, x='inches', color_discrete_sequence=["red"])
fig.update_layout(title='Plotly Histogram', xaxis_title='Height in Inches', yaxis_title='Frequency', width=800, height=600)
# fig.write_image('plotly_histogram.png')
```

```python
sns.displot(df, x='inches', hue='gender', kind='kde', multiple='stack')
```
